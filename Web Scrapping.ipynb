```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.36'}
url="https://www.imdb.com/chart/top/"
response=requests.get(url,headers=headers)
```

```python
soup=BeautifulSoup(response.content,'html.parser')
```

```python
movies_title=[]

for i in soup.select('h3',class_='ipc-title__text'):
    movies_title.append(i.text)
```

```python
data  = {"List Of Movie":movies_title}
df = pd.DataFrame(data)
df = df[1:]
df.to_csv('IMDB_scrapping.csv')
```

```python
url = "https://en.wikipedia.org/wiki/List_of_states_and_territories_of_the_United_States"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
## we need to extract the table
table = soup.find('table', class_='wikitable')
```

```python
for i in table.find_all('tr')[2:]:
    print(i)
    print("----")
```

```python
states=[]

for i in table.find_all('tr')[2:]:
    states.append(i.find('a').text)
```

```python
population=[]

for i in table.find_all('tr')[2:]:
    population.append(i.find('div').text)
```

```python
df = {"States":states,"Population":population}
```

```python
df = pd.DataFrame(df)
df.to_csv('Wikipedia_scapping.csv',index = False)
```
